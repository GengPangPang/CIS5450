### Implementation & Rationale


#### How did you implement this topic and why did you use it?

**[We implemented hypothesis testing prior to model training to evaluate whether individual features or feature groups were statistically associated with hiring outcomes. This allowed us to establish a data-driven foundation for feature selection and better understand the underlying relationships before fitting any models. Depending on variable types, we applied appropriate statistical tests: permutation tests for binary comparisons (e.g., age >35 vs. ≤35, gender), logistic regression for continuous variables (e.g., education level, previous salary), and chi-square tests for categorical associations (e.g., developer status). To further assess feature value, we also used residual-based permutation tests to test whether features like professional experience, education, and salary provided significant additional or joint predictive power. This process ensured that our modeling choices were grounded in statistically validated insights rather than trial-and-error experimentation.]**
